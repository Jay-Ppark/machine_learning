## LeNet 5

### Mnist

```
import tensorflow as tf
import pandas as pd
```

```
# 데이터를 준비 (mnist)
(ind_mdata, dep_mdata), _ = tf.keras.datasets.mnist.load_data()
ind_mdata = ind_mdata.reshape(60000,28,28,1)
dep_mdata = pd.get_dummies(dep_mdata)
print(ind_mdata.shape,dep_mdata.shape)
```

```
# 모델을 생성
X = tf.keras.layers.Input(shape=[28,28,1])

H = tf.keras.layers.Conv2D(6, kernel_size=5, padding='same', activation='swish')(X)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Conv2D(16, kernel_size=5, activation='swish')(H)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(120, activation='swish')(H)
H = tf.keras.layers.Dense(84, activation='swish')(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
model.fit(ind_mdata,dep_mdata,epochs=10)
```

### Cifar 10

```
# 데이터를 준비 (Cifar 10)
(ind_cdata, dep_cdata), _ = tf.keras.datasets.cifar10.load_data()
print(ind_cdata.shape,dep_cdata.shape)
# dep_cdata -> one hot encoding
# get dummies 사용X 이유: dep_cdata가 일차원 형태가 아니기 때문에
dep_cdata = pd.get_dummies(dep_cdata.reshape(50000))
print(ind_cdata.shape,dep_cdata.shape)
```

```
# 모델을 생성
X = tf.keras.layers.Input(shape=[32,32,3])

H = tf.keras.layers.Conv2D(6, kernel_size=5, activation='swish')(X)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Conv2D(16, kernel_size=5, activation='swish')(H)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(120, activation='swish')(H)
H = tf.keras.layers.Dense(84, activation='swish')(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
model.fit(ind_cdata,dep_cdata,epochs=10)
```

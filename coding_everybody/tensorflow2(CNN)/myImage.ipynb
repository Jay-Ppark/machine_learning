```
# 이미지 읽어서 데이터 준비하기
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

paths = glob.glob('./notMNIST_small/*/*.png')
paths = np.random.permutation(paths)
ind_data = np.array([plt.imread(paths[i])for i in range(len(paths))])
dep_data = np.array([paths[i].split('/')[2] for i in range(len(paths))])
print(ind_data.shape,dep_data.shape)
```

```
ind_data = ind_data.reshape(18724,28,28,1)
dep_data = pd.get_dummies(dep_data)
print(ind_data.shape,dep_data.shape)
```

```
import tensorflow as tf

# 모델을 생성
X = tf.keras.layers.Input(shape=[28,28,1])

H = tf.keras.layers.Conv2D(6, kernel_size=5, padding='same', activation='swish')(X)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Conv2D(16, kernel_size=5, activation='swish')(H)
H = tf.keras.layers.MaxPool2D()(H)

H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(120, activation='swish')(H)
H = tf.keras.layers.Dense(84, activation='swish')(H)
Y = tf.keras.layers.Dense(10, activation='softmax')(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
# 모델을 학습
model.fit(ind_data,dep_data,epochs=10)
```

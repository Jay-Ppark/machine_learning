```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

## boston

```
# 데이터 불러오기
bpath = 'https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/boston.csv'
bdata = pd.read_csv(bpath)
print(bdata.columns)
```

```
# 독립변수, 종속변수
ind_bdata = bdata[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
dep_bdata = bdata[['medv']]
print(ind_bdata.shape, dep_bdata.shape)
```

```
# 모델의 구조를 만듬
# H -> hidden layer
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
# 모델의 모양 해석 가능
model.summary()
```

```
# 모델을 학습
# 원래는 25가 한계였는데 multi layer로 정확도가 더 높아짐
model.fit(ind_bdata,dep_bdata,epochs=1000, verbose=0)
model.fit(ind_bdata,dep_bdata,epochs=10)
```

```
# 모델을 이용
print(model.predict(ind_bdata[:5]))
print(dep_bdata[:5])
```

## 아이리스 품종 분류

```
# 데이터를 불러오기
ipath = 'https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/iris.csv'
idata = pd.read_csv(ipath)
print(idata.shape)
```

```
# one hot encoding
idata_enc = pd.get_dummies(idata)
print(idata_enc.columns)
```

```
# 종속변수, 독립변수
ind_idata = idata_enc[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dep_idata = idata_enc[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(ind_idata.shape, dep_idata.shape)
```

```
# 모델의 구조를 만듬
# hidden layer 추가 (H)
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y=tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
# 모델을 학습
model.fit(ind_idata,dep_idata,epochs=100)
```

```
# 모델을 이용
print(model.predict(ind_idata[0:5]))
print(dep_idata[0:5])
```

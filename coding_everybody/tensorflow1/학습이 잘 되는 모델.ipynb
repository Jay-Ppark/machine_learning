## 학습이 잘 되는 모델

* 사용할 레이어
    * tf.keras.layers.BatchNormalization()
    * tf.keras.layers.Activation('swish')

* 데이터
    * 보스턴 집값 예측: https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/boston.csv
    * 아이리스 품종 분류: https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/iris.csv

```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 데이터를 준비
bpath = 'https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/boston.csv'
bdata = pd.read_csv(bpath)
# 종속변수, 독립변수
#print(bdata.columns)
ind_bdata = bdata[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
dep_bdata = bdata[['medv']]
print(ind_bdata.shape, dep_bdata.shape)
```

```
# 모델의 구조를 만듬(구버전)
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(8,activation='swish')(X)
H = tf.keras.layers.Dense(8,activation='swish')(H)
H = tf.keras.layers.Dense(8,activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
# 모델을 학습(구버전)
model.fit(ind_bdata,dep_bdata,epochs=1000)
#loss 15.8358
```

```
# 모델의 구조를 만듬(BatchNormalization)
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
# 모델을 학습(BatchNormalization)
model.fit(ind_bdata,dep_bdata,epochs=1000)
#loss: 8.2063
```

```
# 데이터를 준비
ipath = 'https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/iris.csv'
idata = pd.read_csv(ipath)

# one hot encoding
idata = pd.get_dummies(idata)
print(idata.columns)
# 종속변수, 독립변수
ind_idata = idata[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dep_idata = idata[['품종_setosa', '품종_versicolor','품종_virginica']]
print(ind_idata.shape, dep_idata.shape)
```

```
# 모델의 구조를 만듬(구버전)
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8,activation='swish')(X)
H = tf.keras.layers.Dense(8,activation='swish')(H)
H = tf.keras.layers.Dense(8,activation='swish')(H)
Y = tf.keras.layers.Dense(3,activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',metrics='accuracy')
```

```
# 모델을 학습(구버전)
model.fit(ind_idata,dep_idata,epochs=1000,batch_size=150)
# accuracy: 0.9733
```

```
# 모델의 구조를 만듬(BatchNormalization)
X = tf.keras.layers.Input(shape=[4])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(3,activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',metrics='accuracy')
```

```
# 모델을 학습(BatchNormalization)
model.fit(ind_idata,dep_idata,epochs=1000,batch_size=150)
# accuracy: 1.0000
```

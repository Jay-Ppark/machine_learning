## 보스턴 집값 예측
* github csv url : https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/boston.csv

```
# 14번째 열 MEDV -> 그 town의 주택들 가격의 중앙값
# 1~13 열은 14번째 열에 영향을 미치는 독립변수들
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 데이터를 불러오기
bpath = 'https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/boston.csv'
bdata = pd.read_csv(bpath)
print(bdata.columns)
bdata.head()
```

```
# 독립변수, 종속변수 분리
ind_data = bdata[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
dep_data = bdata[['medv']]
print(ind_data.shape,dep_data.shape)
```

```
# 모델의 구조를 만듬
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
# 데이터로 모델을 학습
model.fit(ind_data,dep_data,epochs=1000, verbose=0)
model.fit(ind_data,dep_data,epochs=10)
```

```
# 모델을 이용
model.predict(ind_data[5:10])
```

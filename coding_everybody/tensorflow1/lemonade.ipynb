## 레모네이드 판매 예측

* github csv url : https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/lemonade.csv


```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 데이터 준비
lpath = 'https://raw.githubusercontent.com/Jay-Ppark/machine_learning/master/coding_everybody/tensorflow1/csv/lemonade.csv'
ldata = pd.read_csv(lpath)
ldata.head()
```

```
# 종속변수, 독립변수
ind_data = ldata[['온도']]
dep_data = ldata[['판매량']]
print(ind_data.shape,dep_data.shape)
```

```
# 모델을 만들기
X=tf.keras.layers.Input(shape=[1])
Y=tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```
# 모델을 학습
#verbose=0 -> 학습하는 과정 출력 x
#model.fit(ind_data,dep_data,epochs=10000, verbose=0)
model.fit(ind_data,dep_data,epochs=10)
```

```
# 모델을 이용
model.predict(ind_data)
```

```
model.predict([[15]])
# 기온이 15일 때 예측 값
```
